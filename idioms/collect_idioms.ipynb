```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
url = "https://toiguru.jp/toeic-phrase-list"
res = requests.get(url)
soup = BeautifulSoup(res.content, "lxml")
# soup
```

```python
tables = soup.find_all("table")[2:]
tables
```

```python
word_dict1 = {}
for table in tables:
    for tr in table.find_all("tr"):
        td_texts = tr.text.split("\n")
        if td_texts[1].replace("\xa0", "") in word_dict1:
            word_dict1[td_texts[1].replace("\xa0", "")] = word_dict1[td_texts[1].replace("\xa0", "")] + "," + td_texts[2]
        word_dict1[td_texts[1].replace("\xa0", "")] = td_texts[2]
word_dict1
```

```python
len(list(word_dict1.values()))
```

```python
url = "https://english.kohsukenemoto.com/toeic/toeic-vocabulary/idioms/"
res = requests.get(url)
soup = BeautifulSoup(res.content, "lxml")
# soup
```

```python
tables = soup.find_all("table")
tables
```

```python
word_dict2 = {}
for table in tables:
    for tr in table.find_all("tr"):
        td_texts = tr.text.split("\n")
        if td_texts[1].replace("\t","") == "英単語":
            continue
        if td_texts[1].replace("\t","") in word_dict2:
            print(td_texts[1].replace("\t",""))
            word_dict2[td_texts[1].replace("\t","")] = word_dict2[td_texts[1].replace("\t","")] + "," + td_texts[2]
        word_dict2[td_texts[1].replace("\t","")] = td_texts[2]
# word_dict2
```

```python
word_dict2["be projected to do"]
```

```python
len(list(word_dict2.values()))
```

```python
word_dict1.update(word_dict2)
```

```python
len(list(word_dict1.values()))
```

```python
word_df = pd.DataFrame(list(word_dict1.items()), columns=["熟語", "意味"])
word_df.head()
```
